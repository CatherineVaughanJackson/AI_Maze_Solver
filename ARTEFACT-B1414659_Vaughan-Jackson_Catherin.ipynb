```python
import tkinter as tk
from functools import partial
import itertools
from itertools import product
import numpy as np
import sys

#import timeit

root = tk.Tk()
window = tk.Frame(root)
window.pack()


#variable size of grid this can be changed
size_of_grid =20
positions = product(range(size_of_grid), range(size_of_grid))

#variable to keep all of the buttons made in the grid so can be accessed by id number
button_ids = []


def highlight(i):
# function to change co-ordinates between wall and space
# black = wall, white = space
   # get the button's identityS
    button_name = (button_ids[i])
    button_colour = button_name.cget('bg')
    if button_colour == "black":
        button_name["bg"] = "white"
    else:
        button_name["bg"]="black"

#create the buttons for the interface
for i in range(size_of_grid):
   # shape the grid
    tk.Canvas(window, width=30, height=0).grid(row=size_of_grid, column=i)
    tk.Canvas(window, width=0, height=30).grid(row=i, column=size_of_grid)

for i, item in enumerate(positions):
    button = tk.Button(window, command=partial(highlight, i), bg = "white")
    button.grid(row=item[0], column=item[1], sticky="nesw")
    button_ids.append(button)



#disable the start and finish points
(button_ids[0])["state"]="disabled"
(button_ids[(size_of_grid**2)-1])["state"]="disabled"



#create button to capture the coordinates
finalise_button = tk.Button(window,text="send", command=lambda:finalise(), bg = "sky blue", width=4, height=2)
finalise_button.grid(column=0, row = (size_of_grid +1))

#set variable to get cells that are "available path"
white_spaces = []

def finalise():
    numbers=[i for i in range(size_of_grid**2)]
    for button in button_ids:
        button_colour = button.cget('bg')
        if button_colour == "white":
            white_spaces.append(button)
            button["text"] = str(numbers[0])
            numbers.pop(0)
    finalise_button["state"]="disabled"
    produce_coord()



#now we have all the buttons that are empty these are our coordinates
#with (0,0) being top left and start point
cell_number = []
coordinates=[]

def produce_coord():
    for entry in white_spaces:
        button_string = str(entry)
        num_cell = button_string.split("n")
        end_num = num_cell[-1]
        cell_number.append(end_num)
       #because the name for button1 and button0 are just labeld as button
       #check status of button individualy and append with new name
       # would have to do the same for (0,0) but since start just leave it
        if cell_number[0] == "":
            cell_number[0] = "1"

#this converts numbers into coordinates top left and bottom right cell will always be there
    for entry in cell_number:
        end_num = int(float(entry))
        column_num =divmod((end_num-1), size_of_grid)
        coordinates.append(column_num)
   #print(coordinates)

    global dictonary
    dictonary= {i: coordinates[i] for i in range(0, len(coordinates))}
    print(dictonary)
    agent_actions()


node_points = []


def agent_actions():
    for key, value in dictonary.items():
       #finds the coords of current index
        x = value[0]
        y =value[1]
        firstkey = key
       #looks right from current  index see if there is a matching coord
        if [key for key, value in dictonary.items() if value == (x +1,y)]:
            right = [key for key, value in dictonary.items() if value == (x +1,y)]
            point = (firstkey, right[0])
            node_points.append(point)
            #print(firstkey, right[0])

       #look left
        if [key for key, value in dictonary.items() if value == (x -1,y)]:
            left = [key for key, value in dictonary.items() if value == (x -1,y)]
            point = (firstkey, left[0])
            node_points.append(point)
            #print(firstkey, left[0])

       #look up
        if [key for key, value in dictonary.items() if value == (x ,y +1)]:
            up = [key for key, value in dictonary.items() if value == (x,y+1)]
            point = (firstkey, up[0])
            node_points.append(point)
            #print(firstkey, up[0])

       #look down
        if [key for key, value in dictonary.items() if value == (x ,y -1)]:
            down = [key for key, value in dictonary.items() if value == (x,y-1)]
            point = (firstkey, down[0])
            node_points.append(point)
            #print(firstkey, down[0])

   #list of points, there are quite a few repeated points
   #print(node_points)

   #list of node then filtered to remove duplicates
    global filtered_nodes
    filtered_nodes = []
    for node in node_points:
        if node not in filtered_nodes and node[::-1] not in filtered_nodes:
            filtered_nodes.append(node)

    print(filtered_nodes)
    
    #now we have the nodes we can set off the q-learning algorithm to find shortet route
    

#SELECT WHICH ALGORITHM TO USE AT THIS POINT 
    
    #Q_learning(filtered_nodes)
    dijkstra_processing(filtered_nodes, dictonary)

    
    
#finally dependant on which algorithm chosen tkinter interface will be updated    
    
def update_tkinter(steps):
   #use white spaces as it'll elliminate some buttons from button_ids
    for button in white_spaces:
        button_text = button.cget("text")
        if int(button_text) in steps:
            button["bg"]="light green"



root.title("Maze")
root.mainloop()
```

```python
def Q_learning(filtered_nodes):
   # goal will always largest number in list
   # because goal point always bottom corner
    
    def max_value(inputlist):
        return max([sublist[-1] for sublist in inputlist])

    goal = max_value(filtered_nodes)

   # function to see if there are any traped points (e.g no connecting adds number
   # to the points list as going to its self in order to not mess up the matrix
    list1 =[i for i in range(goal +1)]
    for point in list1:
        if (point in itertools.chain(*filtered_nodes)):
            pass
        else:
            new_node = (point,point)
            filtered_nodes.append(new_node)

    print(filtered_nodes)

   # matrix size is dependent on number of cells filled here this one is 6
   # because 0,1,2,3,4,5,
   #for this example matrix_size will be one more than goal
    MATRIX_SIZE = goal + 1


    R = np.matrix(np.ones(shape=(MATRIX_SIZE, MATRIX_SIZE)))
    R *= -1


   # assign zeros to paths and 100 to goal-reaching point

    for point in filtered_nodes:
        #print(point)

        if point[1] == goal:
            R[point] = 100
        else:
            R[point] = 0
        if point[0] == goal:
            R[point[::-1]] = 100
        else:
            R[point[::-1]] = 0

   # bottom corner = 100 reward
    R[goal, goal] = 100

    Q = np.matrix(np.zeros([MATRIX_SIZE, MATRIX_SIZE]))

   # learning parameter
    gamma = 0.8
    initial_state = 1


    def available_actions(state):
        current_state_row = R[state,]
        av_act = np.where(current_state_row >= 0)[1]
        return av_act


    available_act = available_actions(initial_state)


    def sample_next_action(available_actions_range):
        next_action = int(np.random.choice(available_act, 1))
        return next_action


    action = sample_next_action(available_act)


    def update(current_state, action, gamma):
        max_index = np.where(Q[action,] == np.max(Q[action,]))[1]

        if max_index.shape[0] > 1:
            max_index = int(np.random.choice(max_index, size=1))
        else:
            max_index = int(max_index)

        max_value = Q[action, max_index]

        Q[current_state, action] = R[current_state, action] + gamma * max_value
       # print('max_value', R[current_state, action] + gamma * max_value)

        if (np.max(Q) > 0):
            return (np.sum(Q / np.max(Q) * 100))
        else:
            return (0)


    update(initial_state, action, gamma)

   # Training
    scores = []
    #for much larger graphs we must increase the range a lot otherwise it shortest route won't be
    
    #CHANGE ITTERATION NUMBER FOR LARGER MAZES
    for i in range(30000):
        current_state = np.random.randint(0, int(Q.shape[0]))
        available_act = available_actions(current_state)

        action = sample_next_action(available_act)
        score = update(current_state, action, gamma)
        scores.append(score)

   # Testing
    current_state = 0
    steps = [current_state]


    while current_state != goal:
        next_step_index = np.where(Q[current_state,] == np.max(Q[current_state,]))[1]

        if next_step_index.shape[0] > 1:
            next_step_index = int(np.random.choice(next_step_index, size=1))
        else:
            next_step_index = int(next_step_index)

        steps.append(next_step_index)
        current_state = next_step_index

    print("Most efficient path:", steps)

    update_tkinter(steps)
```

```python
class Graph(object):
    def __init__(self, nodes, init_graph):
        self.nodes = nodes
        self.graph = self.construct_graph(nodes, init_graph)
        
    def construct_graph(self, nodes, init_graph):
        graph = {}
        for node in nodes:
            graph[node] = {}
        
        graph.update(init_graph)
        
        for node, edges in graph.items():
            for adjacent_node, value in edges.items():
                if graph[adjacent_node].get(node, False) == False:
                    graph[adjacent_node][node] = value
                    
        return graph
    
    def get_nodes(self):
        "Returns the nodes of the graph."
        return self.nodes
    
    def get_outgoing_edges(self, node):
        "Returns the neighbors of a node."
        connections = []
        for out_node in self.nodes:
            if self.graph[node].get(out_node, False) != False:
                connections.append(out_node)
        return connections
    
    def value(self, node1, node2):
        "Returns the value of an edge between two nodes."
        return self.graph[node1][node2]

def dijkstra_algorithm(graph, start_node):
    unvisited_nodes = list(graph.get_nodes())
 
    #  save the cost of visiting each node and update it    
    shortest_path = {}
 
    #  save the shortest known path to a node found so far
    previous_nodes = {}
 
    # e max_value to initialize the "infinity" value of the unvisited nodes   
    max_value = sys.maxsize
    for node in unvisited_nodes:
        shortest_path[node] = max_value
    # However, we initialize the starting node's value with 0   
    shortest_path[start_node] = 0
    
    # The algorithm executes until we visit all nodes
    while unvisited_nodes:
        
        # The code block below finds the node with the lowest score
        current_min_node = None
        for node in unvisited_nodes: 
            if current_min_node == None:
                current_min_node = node
            elif shortest_path[node] < shortest_path[current_min_node]:
                current_min_node = node
                
        # The code block below retrieves the current node's neighbors and updates their distances
        neighbors = graph.get_outgoing_edges(current_min_node)
        for neighbor in neighbors:
            tentative_value = shortest_path[current_min_node] + graph.value(current_min_node, neighbor)
            if tentative_value < shortest_path[neighbor]:
                shortest_path[neighbor] = tentative_value
                # We also update the best path to the current node
                previous_nodes[neighbor] = current_min_node
 
        # After visiting its neighbors node is visited
        unvisited_nodes.remove(current_min_node)
    
    return previous_nodes, shortest_path


def print_result(previous_nodes, shortest_path, start_node, target_node):

    path = []
    node = target_node
    
    while node != start_node:
        path.append(node)
        node = previous_nodes[node]
 
    # Add the start node manually
    path.append(start_node)
    
    #print("".join(reversed(str(path))))
    print(str(path))
    

    update_tkinter(path)
    
```

```python
def dijkstra_processing(filtered_nodes, dictonary):
    

    nodes = []
    for key in dictonary.items():
        nodes.append(key[0])

    init_graph = {}
    for node in nodes:
        init_graph[node] = {}


    for node in filtered_nodes:
        init_graph[node[0]][node[1]]=1


    graph = Graph(nodes, init_graph)
    
    #start node will always be 0
    previous_nodes, shortest_path = dijkstra_algorithm(graph=graph, start_node=0)
    
    
    #find how many nodes there are, as the exit is the bottom pick max value as target_node
    def max_value(inputlist):
        return max([sublist[-1] for sublist in inputlist])

    print_result(previous_nodes, shortest_path, start_node=0, target_node=max_value(filtered_nodes))
    
    
```
